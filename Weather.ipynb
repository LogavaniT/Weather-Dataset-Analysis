```python
data_set['Date/Time'] = pd.to_datetime(data_set['Date/Time'])

# Plotting temperature over time
plt.figure(figsize=(10, 6))
plt.plot(data_set['Date/Time'], data_set['Temp_C'], color='blue')
plt.title('Temperature Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
